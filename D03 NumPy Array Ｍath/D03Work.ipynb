```python
import numpy as np 
```

```python
#1.正常的談話的聲壓為20000微巴斯卡，請問多少分貝?
#請寫下程式
V1 = 20000
V0 = 20

GdB = 20 * np.log10(V1/V0)
##答案
GdB
```

```python
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
# GdB / 20 = np.log10(V1/V0)
# 10 ** (GdB / 20) = V1 / V0
# V1 = V0 * (10 ** (GdB / 20))
# [指數對數方程式](https://youtu.be/VdlfHl-PBLI)
def Ｖ1Fn(Gdb): 
    return V0 * (10**(Gdb/20))

V30 = Ｖ1Fn(30)
V50 = Ｖ1Fn(50)
##答案
V30/V50
```
