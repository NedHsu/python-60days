題目:<br>
1. 將下兩列array存成npz檔<br>
array1 = np.array(range(30))<br>

2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔


```python
import numpy as np
```

```python
#. 將下兩列array存成npz檔
array1 = np.array(range(30))
array2 = np.array([2,3,5])
filename1 = 'work01.npz'
filename2 = 'work02.npz'

with open(filename1, 'wb') as f:
    np.savez(f, array1 = array1, array2 = array2)
```

```python
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔
array3 = np.array([5,4,3])

npzfile = np.load(filename1)
with open(filename2, 'wb') as f:
    np.savez(f, array1 = npzfile['array1'], array2 = npzfile['array2'], array3 = array3)

#Check
npzfile2 = np.load(filename2)
for file in npzfile2.files:
    print(npzfile2[file])
```
